Import pandas and sklearn

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

Read the dataset

```python
data = pd.read_csv('/content/toys.csv')
data
```

Display the top 5 observations in the dataset

```python
data.head()
```

compute the count of unique values in the column named cough

```python
data['cough'].value_counts()
```

```python
#mild
#strong
```

```python
data_Ordinal=data[['cough']]
data_Ordinal.head()
```

```python
oe = OrdinalEncoder(categories=[["Mild", "Strong"]])
```

```python
oe.fit(data_Ordinal)
```

```python
data_Transform =oe.transform(data_Ordinal)
```

```python
data_Transform
```

```python
New_data = pd.DataFrame(data_Transform, columns=['cough'])
```
